```
import numpy as np #calcolo matriciale
import tensorflow as tf #reti neurali
import matplotlib.pyplot as plt #grafici
```

```
#creo dataset con 200 righe 1 colonna
np.random.seed(42) #per vedere i risultati
x = np.random.standard_normal((200,1))
```

```
#dataset di 200 righe con valori casuali 
x
```

```
#creo Y
y = 2 * x + 3
```

```
#mostra y
y
```

```
# creo grafico di x rispetto y
plt.plot(x,y)
plt.grid(True)#aggiunge griglia 
```

```
#definisco modello sequenziale 
model = tf.keras.Sequential([
                             tf.keras.layers.Dense(1,input_shape=[1])
]) #modello da 1 layer cioé un neurone, input shape num colonne

model.summary() #crea un sommario
```

```
#rapresentazioe grafica modello
tf.keras.utils.plot_model(model,show_shapes=True)
```

```
#compilazione modello
model.compile(loss=tf.keras.losses.mean_squared_error,optimizer=tf.keras.optimizers.Adam())
```

```
#fit modello(training modello)
history = model.fit(x = x, y=y, epochs=1000) #epochs = quante volte processo parametri
```

```
# risultati predizioni
y_pred = model.predict(x)
```

```
#confronto con y originali
np.c_[y,y_pred]
```

```
#loss totale
tf.keras.losses.mean_squared_error(y[:,0],y_pred[:,0])
```

```
#grafico loss nel tempo
loss_hist = history.history['loss']
plt.plot(range(1000),loss_hist)
plt.grid(True)
```
